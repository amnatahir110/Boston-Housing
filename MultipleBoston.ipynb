```python
import numpy as np
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
```

```python
#load dataset
df=pd.read_csv('C:\\Users\\cvalley\\.jupyter\\MultipleBoston.csv')
df.head()
```

```python
df.nox.median()
```

```python
import math
median_nox = math.floor(df.nox.median())
median_nox
```

```python
df.nox.fillna(median_nox)
```

```python
df.nox=df.nox.fillna(median_nox)
df
```

```python
lin=linear_model.LinearRegression()
lin.fit(df[['crim','zn','indus','chas','nox','rm','age','dis','rad','tax','ptratio','b','lstat']],df.medv)
```

```python
m_coef=lin.coef_
m_coef
```

```python
intercept=lin.intercept_
intercept
```

```python
medv= m_coef[0]* df['crim'] + m_coef[1]*df['zn'] + m_coef[2]*df['indus'] + m_coef[3] *df['chas']  + m_coef[4]*df['nox'] + m_coef[5]*df['rm'] + m_coef[6]*df['age'] + m_coef[7]*df['dis'] + m_coef[8]* df['rad'] + m_coef[9]* df['tax']  + m_coef[10]*df['ptratio'] + m_coef[11]*df['b'] + m_coef[12]*df['lstat'] + intercept
print(medv)
```
